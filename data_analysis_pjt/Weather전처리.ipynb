```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'AppleGothic'
import warnings
warnings.filterwarnings('ignore')

df_temp = pd.read_excel('./Downloads/BigDataPjtData/평균기온.xlsx')
df_rain = pd.read_excel('./Downloads/BigDataPjtData/강수량.xlsx')
df_light = pd.read_excel('./Downloads/BigDataPjtData/일사량.xlsx')
df_wind = pd.read_excel('./Downloads/BigDataPjtData/순간최대풍속.xlsx')
df_humid = pd.read_excel('./Downloads/BigDataPjtData/습도.xlsx')
#평균온도(℃), 강수량(mm), 일사량(MJ/m2), 순간최대풍속(m/s), 습도(%)
```

```python
df_temp['평균온도'] = (df_temp['충청남도'] + df_temp['경기도'] + df_temp['전라남도'] + df_temp['전라북도'])/4
df_rain['강수량'] = (df_rain['충청남도'] + df_rain['경기도'] + df_rain['전라남도'] + df_rain['전라북도'])/4
df_light['일사량'] = (df_light['충청남도'] + df_light['경기도'] + df_light['전라남도'] + df_light['전라북도'])/4
df_wind['순간최대풍속'] = (df_wind['충청남도'] + df_wind['경기도'] + df_wind['전라남도'] + df_wind['전라북도'])/4
df_humid['습도'] = (df_humid['충청남도'] + df_humid['경기도'] + df_humid['전라남도'] + df_humid['전라북도'])/4
df_humid.head()
```

```python
df_temp['년'] = df_temp.iloc[:, 0].str[:4]
df_temp['월'] = df_temp.iloc[:, 0].str[5:]
df_temp.head()
```

```python
df = pd.concat([df_temp.iloc[:, [6, 7, 5]], df_rain.iloc[:, 5], df_light.iloc[:, 5], df_wind.iloc[:, 5], df_humid.iloc[:, 5]], axis=1)
df = round(df, 2)
df
```

```python
df.to_csv('Downloads/BigDataPjtData/weather_final.csv')
```
